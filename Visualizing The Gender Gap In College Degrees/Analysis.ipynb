```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')#read in file

#different colour lines to differentiate between men and women
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

#list of subjects that belong to each category
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

#set plot size
fig = plt.figure(figsize=(10, 18))

#plot categories per column
pos = 1
for sp in range(0,6):
    ax = fig.add_subplot(6,3,pos)#place plot in correct order and position
    
    #plot both statistics for men and women from each category 
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    
    #remove spine from plots
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    #set plot aesthetics
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp], y = 1.1)
    ax.tick_params(bottom='off', top='off', left='off', right='off', labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    
    #only include line labels on first and last plots per column to reduce clutter
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
    pos+=3

pos = 2
for sp in range(0,5):
    ax = fig.add_subplot(6,3,pos)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[sp], y = 1.1)
    ax.tick_params(bottom='off', top='off', left='off', right='off', labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 4:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
    pos+=3

    
pos = 3
for sp in range(0,6):
    ax = fig.add_subplot(6,3,pos)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[sp], y = 1.1)
    ax.tick_params(bottom='off', top='off', left='off', right='off', labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
    pos+=3

#set spacing between plots for readability
fig.subplots_adjust(hspace=.3)
fig.subplots_adjust(wspace=.3)

plt.savefig('gender_degrees.png')#save plots to image file for easier access
plt.show()
```
